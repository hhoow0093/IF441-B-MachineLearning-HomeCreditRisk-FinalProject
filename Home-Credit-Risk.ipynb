### Data cleaning

#### Bagi data numerical, categorical, dan boolean

```python
from pandas.api.types import is_numeric_dtype, is_bool_dtype, is_string_dtype, is_categorical_dtype

def is_numeric_boolean(feature):
  unique_value = feature.dropna().unique()
  return set(unique_value).issubset({0, 1})


def categorize_features(df, id_threshold= 0.95, cat_unique_thres=0.05):
  n = len(df)
  result = {
      "numeric" : [],
      "boolean" : [],
      "categorical": [],
      "identifier": []
  }
  for col in df.columns:
    feature = df[col]
    dtype = feature.dtype

    if is_bool_dtype(feature) or is_numeric_boolean(feature):
      result["boolean"].append(col)

    elif is_numeric_dtype(feature):
      num_unique = feature.nunique(dropna=True)
      if num_unique / n > id_threshold:
        result["identifier"].append(col)
      else:
        result["numeric"].append(col)

    elif is_categorical_dtype(feature) or is_string_dtype(feature):
      num_unique = feature.nunique(dropna=True)
      if num_unique / n > id_threshold:
        result["identifier"].append(col)
      elif num_unique / n < cat_unique_thres:
        result["categorical"].append(col)
      else:
        result["categorical"].append(col)
  return result
```

```python
dictionary = categorize_features(X)
for key, cols in dictionary.items():
    print(f"\n{key.upper()} ({len(cols)} features):")
    for col in cols:
        print(f"  - {col}")
```

```python
import numpy as np
upper = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

threshold = 0.8
high_corr = [
    (col, row, corr_matrix.loc[row, col])
    for col in upper.columns
    for row in upper.index
    if not pd.isnull(upper.loc[row, col]) and abs(upper.loc[row, col]) > threshold
]

print("Highly correlated feature pairs (|r| > 0.8):\n")
for col1, col2, corr in high_corr:
    print(f"{col1} - {col2}: correlation = {corr:.2f}")
```

#### Fitur ekstraksi

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np

pca_models = {}

for col1, col2, _ in high_corr:
    if col1 == col2:
        continue

    pair = [col1, col2]
    new_col = f'PCA_{col1}_{col2}'

    if all(c in df_clean_numerical_train.columns for c in pair):
        scaler = StandardScaler()
        scaled_train = scaler.fit_transform(df_clean_numerical_train[pair])

        pca = PCA(n_components=1)
        pca_train = pca.fit_transform(scaled_train)
        df_clean_numerical_train[new_col] = pca_train

        missing_in_test = [c for c in pair if c not in df_clean_numerical_test.columns]
        for c in missing_in_test:
            df_clean_numerical_test[c] = np.nan  # add missing columns

        temp_test = df_clean_numerical_test[pair].fillna(df_clean_numerical_test[pair].mean())

        scaled_test = scaler.transform(temp_test)
        df_clean_numerical_test[new_col] = pca.transform(scaled_test)


        df_clean_numerical_train = df_clean_numerical_train.drop(columns=pair, errors='ignore')
        df_clean_numerical_test = df_clean_numerical_test.drop(columns=pair, errors='ignore')

        pca_models[new_col] = (scaler, pca)

        print(f"✅ Created PCA feature: {new_col}")


```
